```python

#functions =  are blocks of code within a script or algorithm that perform a specific task. 
#There are four basic parts to a function:
# Name
# Parameters
# Code block
# Return value

    # x = 'global value'

    # def foo():
    #     x = 'local value'
    #     print(x)

    # foo()
    # print(x)
# output
    # local value
    # global value

#movie is parameter
#movies are dicts and we want to make nondestructive edits
### WHAT IS MOVIE? 
def clean_movie(movie):
    #create a new LOCAL variable called movie, and assign it the new copy of the parameter "movie"
        #inside this function, "movie" refers to the local copy, changes will refer to only this copy
    movie = dict(movie) # = CONSTRUCTOR -> function that inicializes new objects:
                        #                 (reserve space in memory and copies all info into the space)
                        #This makes a copy dict() builds dictrionary directly from sequences of key-value pairs

    # make an empty dict to hold alternative titles
    alt_titles = {}

    #The different language columns are for alternate titles of the movie. 
    # Need to combine all of them into one dictionary that has all the alternate titles. 
    #loop through a list of all alternative title keys - columns are alternative title keys
    #Check if the current key exists in the movie object.
    #this is a list of all the columns that we don't think are helpful
    # KEY = COLUMN
    for key in ['Also known as','Arabic','Cantonese','Chinese','French',
                'Hangul','Hebrew','Hepburn','Japanese','Literally',
                'Mandarin','McCune–Reischauer','Original title','Polish',
                'Revised Romanization','Romanized','Russian',
                'Simplified','Traditional','Yiddish']:
         #If the current key is in movie, remove the key-value pair and add to the alternative titles dict.
        if key in movie:
            #put that key/value pair into the alternative titles dictionary
            alt_titles[key] = movie[key]
            #print(key)
           
            #The pop() method removes the item at the given index from the list and returns the removed item.
            #list_name.pop(index)
            #Remove from movie bc it is in alternative titles
            movie.pop(key)
    #After looping through every key, add the alternative titles dict to the movie object
    # if there is anything in the alt_titles dictionary (key = arabic, cantonese, etc, value = whatevs in that column)
    if len(alt_titles) > 0:
        # then put it all into a column/dict  
        movie['alt_titles'] = alt_titles
        #print(movie['alt_titles'])

    #consolidate columns with the same data
    # merge column names
    def change_column_name(old_name, new_name):
        #check if the key (column) exists in the record
        # if the old/first name is in there, then...
        if old_name in movie:
            # make it into the new name, and drop the old name (rename it)
            movie[new_name] = movie.pop(old_name)
    change_column_name('Adaptation by', 'Writer(s)')
    change_column_name('Country of origin', 'Country')
    change_column_name('Directed by', 'Director')
    change_column_name('Distributed by', 'Distributor')
    change_column_name('Edited by', 'Editor(s)')
    change_column_name('Length', 'Running time')
    change_column_name('Original release', 'Release date')
    change_column_name('Music by', 'Composer(s)')
    change_column_name('Produced by', 'Producer(s)')
    change_column_name('Producer', 'Producer(s)')
    change_column_name('Productioncompanies ', 'Production company(s)')
    change_column_name('Productioncompany ', 'Production company(s)')
    change_column_name('Released', 'Release Date')
    change_column_name('Release Date', 'Release date')
    change_column_name('Screen story by', 'Writer(s)')
    change_column_name('Screenplay by', 'Writer(s)')
    change_column_name('Story by', 'Writer(s)')
    change_column_name('Theme music composer', 'Composer(s)')
    change_column_name('Written by', 'Writer(s)')

    return movie
```

```python
#make a function for how we're going to get all the info from this column into the same format

def parse_dollars(s):
    
    
#     we made all the applicable fields into strings earlier
    # if s is not a string, return NaN
    if type(s) != str:
        return np.nan

    # if input is of the form $###.# million
#         re.match(pattern, string) see if our string(s - from parameter we're passing through) matches a pattern
    if re.match(r'\$\s*\d+\.?\d*\s*milli?on', s, flags=re.IGNORECASE):

        # remove dollar sign and " million"
#             re.sub(pattern = (find this), replacement_string = (replace w/ this), string (this is what we're looking through))
#                     find this expression in the string and replace it w/ a space
        s = re.sub('\$|\s|[a-zA-Z]','', s)

        # convert to float and multiply by a million
        value = float(s) * 10**6

        # return value
        return value

    # if input is of the form $###.# billion
    elif re.match(r'\$\s*\d+\.?\d*\s*billi?on', s, flags=re.IGNORECASE):

        # remove dollar sign and " billion"
        s = re.sub('\$|\s|[a-zA-Z]','', s)

        # convert to float and multiply by a billion
        value = float(s) * 10**9

        # return value
        return value

    # if input is of the form $###,###,###
    elif re.match(r'\$\s*\d{1,3}(?:[,\.]\d{3})+(?!\s[mb]illion)', s, flags=re.IGNORECASE):

        # remove dollar sign and commas
        s = re.sub('\$|,','', s)

        # convert to float
        value = float(s)

        # return value
        return value

    # otherwise, return NaN
    else:
        return np.nan
```

```python
#write a fn that fills in missing data and drops redundant column

def fill_missing_kaggle_data(df, kaggle_column, wiki_column):
    df[kaggle_column] = df.apply(
        lambda row: row[wiki_column] if row[kaggle_column] == 0 else row[kaggle_column]
        , axis=1)
    df.drop(columns=wiki_column, inplace=True)
```
